```python
# ðŸŽ“ Student Grade Calculator ðŸŽ“

print("Welcome to the Student Grade Calculator!")
print("----------------------------------------")

# Step 1: Get marks from the user
marks = float(input("Enter your marks (0-100): "))

# Step 2: Validate the input
if marks < 0 or marks > 100:
    print("âŒ Invalid input! Please enter marks between 0 and 100.")
else:
    # Step 3: Use if-else to find the grade
    if marks >= 90:
        grade = "A"
        message = "Excellent work! ðŸŒŸ"
    elif marks >= 80:
        grade = "B"
        message = "Great job! Keep it up! ðŸ‘"
    elif marks >= 70:
        grade = "C"
        message = "Good effort! You can do even better! ðŸ’ª"
    elif marks >= 60:
        grade = "D"
        message = "You passed, but aim higher next time! ðŸŽ¯"
    else:
        grade = "F"
        message = "Donâ€™t give up â€” keep practicing! ðŸ“˜"

    # Step 4: Show the result
    print(f"\nYour Grade: {grade}")
    print(f"Feedback: {message}")
```

```python
import pandas as pd
from datetime import datetime
import matplotlib.pyplot as plt
```

```python
data = {
    "OrderID": [1001,1002,1003,1004,1005],
    "Date": ["2025-01-05","2025-01-12","2025-01-20","2025-02-01","2025-02-03"],
    "Product": ["Notebook","Pen","Notebook","Stapler","Pen"],
    "UnitPrice": [3.50,0.75,3.50,7.00,0.75],
    "Quantity": [10,50,5,3,40]
}
```

```python
import pandas as pd   # make sure pandas is imported first
df = pd.DataFrame(data)
```

```python
print(df)
```

```python
df['Date'] = pd.to_datetime(df['Date'])
print(df)
```

```python
df['Sales'] = df['UnitPrice'] * df['Quantity']
print(df)
```

```python
total_sales = df['Sales'].sum()
print("Total Sales:", total_sales)
```

```python
top_products = df.groupby('Product', as_index=False)['Sales'].sum()
print(top_products)
```

```python
top_products = top_products.sort_values('Sales', ascending=False)
print(top_products)
```

```python
df['YearMonth'] = df['Date'].dt.to_period('M').astype(str)
monthly_sales = df.groupby('YearMonth', as_index=False)['Sales'].sum()
print(monthly_sales)
```

```python
import matplotlib.pyplot as plt

plt.bar(top_products['Product'], top_products['Sales'])
plt.title('Sales by Product')
plt.xlabel('Product')
plt.ylabel('Sales')
plt.show()
```

```python
plt.plot(monthly_sales['YearMonth'], monthly_sales['Sales'], marker='o')
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.grid(True)
plt.show()
```

```python
print("Total Sales:", total_sales)
print("Best Product:", top_products.iloc[0]['Product'])
```

```python
# Step 1: Install and import required libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Step 2: Create sample weather data
data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
              'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'],
    'Avg_Temperature': [15, 17, 22, 28, 33, 36, 35, 34, 30, 25, 20, 16],
    'Rainfall_mm': [120, 100, 80, 40, 20, 10, 5, 10, 30, 60, 90, 110],
    'Humidity_%': [75, 70, 65, 55, 50, 45, 40, 45, 55, 65, 70, 75]
}

df = pd.DataFrame(data)
df
```

```python
# Step 3: Explore the data
print(df.info())
print(df.describe())
```

```python
# Step 4a: Line chart for temperature
plt.figure(figsize=(8,5))
plt.plot(df['Month'], df['Avg_Temperature'], marker='o', color='orange')
plt.title('Average Monthly Temperature')
plt.xlabel('Month')
plt.ylabel('Temperature (Â°C)')
plt.grid(True)
plt.show()
```

```python
# Step 4b: Bar chart for rainfall
plt.figure(figsize=(8,5))
plt.bar(df['Month'], df['Rainfall_mm'], color='skyblue')
plt.title('Monthly Rainfall (mm)')
plt.xlabel('Month')
plt.ylabel('Rainfall (mm)')
plt.show()
```

```python
# Step 4c: Pie chart for humidity
plt.figure(figsize=(6,6))
plt.pie(df['Humidity_%'], labels=df['Month'], autopct='%1.1f%%', startangle=90)
plt.title('Humidity Distribution by Month')
plt.show()
```

```python
print("ðŸŒ¦ï¸ WEATHER DATA ANALYSIS REPORT ðŸŒ¦ï¸\n")

# Key insights
print("1ï¸âƒ£ The hottest month is:", df.loc[df['Avg_Temperature'].idxmax(), 'Month'])
print("2ï¸âƒ£ The rainiest month is:", df.loc[df['Rainfall_mm'].idxmax(), 'Month'])
print("3ï¸âƒ£ The most humid month is:", df.loc[df['Humidity_%'].idxmax(), 'Month'])

# Show charts again
plt.figure(figsize=(8,5))
plt.plot(df['Month'], df['Avg_Temperature'], marker='o', color='orange')
plt.title('Average Monthly Temperature')
plt.show()

plt.figure(figsize=(8,5))
plt.bar(df['Month'], df['Rainfall_mm'], color='skyblue')
plt.title('Monthly Rainfall (mm)')
plt.show()

plt.figure(figsize=(6,6))
plt.pie(df['Humidity_%'], labels=df['Month'], autopct='%1.1f%%', startangle=90)
plt.title('Humidity Distribution by Month')
plt.show()
```
